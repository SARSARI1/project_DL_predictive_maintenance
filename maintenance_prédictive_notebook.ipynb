- **Machine failure :** Indicateur binaire (0 ou 1) représentant si une panne de la machine s'est produite.
- **TWF :** Indicateur de défaillance d'usure de l'outil (0 ou 1).
- **HDF :** Indicateur de défaillance de dissipation de chaleur (0 ou 1).
- **PWF :** Indicateur de défaillance de puissance (0 ou 1).
- **OSF :** Indicateur de défaillance de surcharge (0 ou 1).
- **RNF :** Indicateur de défaillance aléatoire (0 ou 1).

# <a> Analysez une variable quantitative et une qualitative par ANOVA</a>

Voici le code qui permet de représenter une variable quantitative et une variable qualitative. Tout d'abord, créez le sous-échantillon sur lequel vous souhaitez travailler en adaptant ce code, notamment les variables X et Y selon la question que vous aurez choisie parmi celles ci-dessus

```python
import pandas as pd
import matplotlib.pyplot as plt

# Définir les noms des variables
X = "Type" # Variable qualitative
Y = "Air temperature [K]" # Variable quantitative

# Définir les types (modalités) que vous voulez tracer
types = data[X].unique()  # ["M", "L", "H"] si ce sont vos types dans la colonne "Type"

# Préparer les données pour le boxplot
boxplot_data = [data[data[X] == test][Y] for test in types]

# Définir les étiquettes pour le boxplot
labels = types

# Propriétés graphiques
medianprops = {'color': "black"}
meanprops = {'marker': 'o', 'markeredgecolor': 'black', 'markerfacecolor': 'firebrick'}

# Créer le boxplot
plt.boxplot(boxplot_data, labels=labels, showfliers=False, medianprops=medianprops, 
            vert=False, patch_artist=True, showmeans=True, meanprops=meanprops)

# Définir les étiquettes et le titre
plt.xlabel(X)
plt.ylabel(Y)
plt.title("Évolution de la température en fonction du type de produit")

# Afficher le graphique
plt.show()

# Sous-échantillon pour vérification
sous_echantillon = data[Y]
print(sous_echantillon.head())
```

# <a > Creation et Aprentissage du modele :</a>

Encodage des variables qualitatives  :

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
data['Product ID'] = encoder.fit_transform(data['Product ID'])
data['Type'] = encoder.fit_transform(data['Type'])
```

```python
data.head()
```

**Training :**

```python
x=data[['UDI', 'Product ID', 'Type', 'Air temperature [K]',
       'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]',
       'Tool wear [min]']]
```

```python
y=data[['Machine failure']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8 )
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
from sklearn.ensemble import RandomForestRegressor


model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
nb.fit(x_train,y_train['Machine failure'].values)
```

```python
nb.score(x_test,y_test['Machine failure'].values)
```

```python
from sklearn.metrics import confusion_matrix, f1_score, recall_score, accuracy_score, precision_score
```

```python
y_pred = nb.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print("F1_score :", f1_score(y_pred,y_test))
print("Accuracy :", accuracy_score(y_pred,y_test))
print("Recall :", recall_score(y_pred,y_test))
print("Precision: ", precision_score(y_pred,y_test))
```

**Test:**

```python
def predictMaintenace(pre):
    Target =nb.predict(pre)
    if(Target[0] ==0):
        print("machine needs repair")
    else:
         print("no need for repair")
```
